```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread("test.png")

cv2_imshow(image)
```

```python

def apply_color_balance(image, r_factor, g_factor, b_factor):

  b, g, r = cv2.split(image)

  r = cv2.convertScaleAbs(r, alpha=r_factor)
  g = cv2.convertScaleAbs(g, alpha=g_factor)
  b = cv2.convertScaleAbs(b, alpha=b_factor)

  balanced_image = cv2.merge((b, g, r))
  return balanced_image
```

```python
r_fac = 1.5
g_fac = 1.5
b_fac = 1.5

balanced_image = apply_color_balance(image, r_fac, g_fac, b_fac)

cv2_imshow(balanced_image)
```
