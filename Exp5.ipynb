```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv.imread('wallpaper.jpg',cv.IMREAD_GRAYSCALE)
```

```python
def plot_histogram_cdf(image):
  hist , bins = np.histogram(image.flatten(),256,[0,256])
  cdf = hist.cumsum()
  cdf_normalized = cdf * hist.max()/ cdf.max()

  plt.plot(cdf_normalized, color = 'b')
  plt.hist(image.flatten(),256,[0,256], color = 'r')
  plt.xlim([0,256])
  plt.legend(('cdf','histogram'), loc = 'upper left')
  plt.show()
```

```python
def full_historgram_equalization(image):
  image_equalized = cv.equalizeHist(image)
  return image_equalized
```

```python
def partial_histogram_equalization(image):
  clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
  image_equalized = clahe.apply(image)
  return image_equalized
```

```python
plot_histogram_cdf(image)
```

```python
plot_histogram_cdf(image)

# Example usage of the other functions:
equalized_image_full = full_historgram_equalization(image)
cv2_imshow(equalized_image_full)
plot_histogram_cdf(equalized_image_full)


equalized_image_partial = partial_histogram_equalization(image)
cv2_imshow(equalized_image_partial)
plot_histogram_cdf(equalized_image_partial)
```
