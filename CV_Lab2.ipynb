```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image1 = cv2.imread("test.png")

cv2_imshow(image1)
```

```python





brightness = 50

img = cv2.add(image1, brightness)

cv2_imshow(img)
```

```python
contrast = 1.8

imgconst = cv2.convertScaleAbs(image1, alpha=contrast, beta=-50)
# note : plz adjuct the the beta if you want image contract to be dark then beat negative
cv2_imshow(imgconst)
```

```python
th_value = 110

ret, th = cv2.threshold(image1, th_value, 255, cv2.THRESH_BINARY)

cv2_imshow(th)
```
